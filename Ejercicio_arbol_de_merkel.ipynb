```python
import hashlib                                    # Librería para generar funciones hash criptográficas

def hash(text, hasharr):                          # Función para generar el hash de un texto y guardarlo
    hash_object = hashlib.sha256(text.encode())   # Creamos el objeto hash usando el algoritmo SHA-256, encode() convierte el texto a bytes
    hex_dig = hash_object.hexdigest()             # Convertimos el hash a su representación hexadecimal (cadena de 64 caracteres)
    hasharr.append(hex_dig)                       # Agregamos el hash generado a la lista de hashes
    print(hex_dig)                                # Imprimimos el hash generado
    return hasharr                                # Retornamos la lista actualizada con el nuevo hash

def conc(hasharr):                                # Función para concatenar hashes y generar nuevos hashes a partir de ellos
    conc = hasharr[0]                             # Tomamos el primer hash de la lista
    con2 = hasharr[1]                             # Tomamos el segundo hash de la lista
    conc3 = conc + con2                           # Concatenamos el primer y segundo hash
    conc4 = hasharr[2]                            # Tomamos el tercer hash de la lista
    conc5 = hasharr[3]                            # Tomamos el cuarto hash de la lista
    conc6 = conc4 + conc5                         # Concatenamos el tercer y cuarto hash
    text = conc3                                  # Preparamos la primera concatenación para hashearla
    hash(text, hasharr)                           # Generamos el hash de la concatenación 1+2 y lo agregamos a la lista
    text = conc6                                  # Preparamos la segunda concatenación para hashearla
    hash(text, hasharr)                           # Generamos el hash de la concatenación 3+4 y lo agregamos a la lista
    conc7 = hasharr[4]                            # Tomamos el quinto hash (resultado de concatenar 1+2)
    conc8 = hasharr[5]                            # Tomamos el sexto hash (resultado de concatenar 3+4)
    conc9 = conc7 + conc8                         # Concatenamos los dos hashes intermedios
    text = conc9                                  # Preparamos la concatenación final para hashearla
    hash(text, hasharr)                           # Generamos el hash raíz a partir de todos los anteriores
```

```python
import hashlib                                                                  # Librería para generar funciones hash criptográficas

class node:                                                                     # Clase que representa un nodo del árbol de Merkle
    def __init__(self, value, left=None, right=None):                           # Constructor: recibe el valor del hash y sus hijos opcionales
        self.value = value                                                      # Hash del nodo
        self.left = left                                                        # Puntero al hijo izquierdo (None si es hoja)
        self.right = right                                                      # Puntero al hijo derecho (None si es hoja)

def hash(n):                                                                    # Función para generar el hash SHA-256 de cualquier valor
    hash_object = hashlib.sha256(str(n).encode())                               # Convertimos el valor a texto, luego a bytes y aplicamos SHA-256
    hex_dig = hash_object.hexdigest()                                           # Obtenemos la representación hexadecimal del hash (64 caracteres)
    return hex_dig                                                              # Retornamos el hash generado

def conc(n1, n2):                                                               # Función para concatenar dos hashes
    n3 = n1 + n2                                                                # Unimos los dos hashes en uno solo
    return n3                                                                   # Retornamos la concatenación

def tree(objects_in_use):                                                       # Función principal que construye el árbol de Merkle completo
    actual_lev = [node(hash(n)) for n in objects_in_use]                        # Creamos las hojas: un nodo con el hash de cada elemento
    next_lev = []                                                               # Lista auxiliar para construir el siguiente nivel
    while len(actual_lev) > 1:                                                  # Mientras haya más de un nodo en el nivel actual, seguimos subiendo
        if len(actual_lev) % 2 != 0:                                            # Si el nivel tiene cantidad impar de nodos...
            actual_lev.append(actual_lev[-1])                                   # ...duplicamos el último nodo para poder emparejarlo
        for i in range(0, len(actual_lev), 2):                                  # Recorremos los nodos de dos en dos para emparejarlos
            left = actual_lev[i].value                                          # Tomamos el hash del nodo izquierdo del par
            right = actual_lev[i+1].value                                       # Tomamos el hash del nodo derecho del par
            parent = hash(conc(left, right))                                    # Generamos el hash padre concatenando los dos hashes hijos
            parent_left = actual_lev[i]                                         # Guardamos el nodo izquierdo como hijo izquierdo del padre
            parent_right = actual_lev[i+1]                                      # Guardamos el nodo derecho como hijo derecho del padre
            next_lev.append(node(parent, parent_left, parent_right))            # Creamos el nodo padre y lo agregamos al siguiente nivel
        actual_lev = next_lev                                                   # El siguiente nivel se convierte en el nivel actual
        next_lev = []                                                           # Vaciamos la lista auxiliar para construir el próximo nivel
    print(actual_lev[0].value)                                                  # Imprimimos el hash raíz (representa todos los datos del árbol)
    return actual_lev[0]                                                        # Retornamos el nodo raíz con toda la estructura del árbol
```
