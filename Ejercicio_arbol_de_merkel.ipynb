```python
import hashlib

class node:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right

def hash(n):
    hash_object = hashlib.sha256(str(n).encode())
    hex_dig = hash_object.hexdigest()
    return hex_dig

def conc(n1, n2):
    n3 = n1 + n2
    return n3

def tree(objects_in_use):                                                       #Objects_in_use es la lista de elementos a utilizar
    actual_lev = [node(hash(n)) for n in objects_in_use]
    next_lev = []
    while len(actual_lev) > 1:
      if len(actual_lev) % 2 != 0:
        actual_lev.append(actual_lev[-1])
      for i in range(0, len(actual_lev), 2):
        left = actual_lev[i].value
        right = actual_lev[i+1].value
        parent = hash(conc(left, right))
        parent_left = actual_lev[i]
        parent_right = actual_lev[i+1]
        next_lev.append(node(parent, parent_left, parent_right))
      actual_lev = next_lev
      next_lev = []
    print (actual_lev[0].value)
    return actual_lev[0]
```
