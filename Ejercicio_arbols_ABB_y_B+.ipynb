```python
import random                   #Importamos la libreria para randomizar numeros
from faker import Faker         #Usamos la libreria faker previamente instalada que se utiliza para randomizar los nombres

def generate_data():            #Funcion que genera los datos en forma de diccionario de forma aleatoria
    fake = Faker('es_ES')       #Carga la herramienta como "faker" para los nombres
    used_ids = set()            #Genera un conjunto en base a hashes mediante una tabla que permite saber que ID ya estan en la lista de forma rapida
    data = []                   #Genera la lista vacia a donde se van a ingrear los datos creados

    while len(data) < 10000:                          #Loop para la creacion de los diccionarios
        new_id = random.randint(0, 99999)             #Genera el id
        if new_id in used_ids:                        #Verifica si el id ya existe
            continue                                  #Si ya existe el id, salta al incio del while para generar otro ID
        used_ids.add(new_id)                          #Agrega el ID al conjunto de IDś
        nombre = fake.first_name()                    #Genera el nombre de forma aleatoria
        promedio = round(random.uniform(0.0, 10.0), 1)                          #Genera el número para el promedi oentre 0.0 y 10.0 de forma aleatoria
        data.append({"id": new_id, "nombre": nombre, "promedio": promedio})     #Agrea a la lista de datos el diccionario con los datos generados actualemnte

    with open("datos.txt", "w", encoding="utf-8") as f:                         #Genera o sobre escribe un documento txt en codigicacion UTF-8 en formato de escritura como f
        for d in data:                                                          #Loop para recorrer la lista de datos
            f.write(f'{{"id": {str(d["id"]).zfill(5)}, "nombre": "{d["nombre"]}", "promedio": {d["promedio"]}}},\n')           #escribe sobre el archivo txt previamente creado cada diccionario con el foramto establecido

    print("Archivo generado con éxito: datos.txt")                              #Imprime un aviso de terminado

generate_data()                 #Ejecuta la función
```

```python
import re                      #Importamos la libreria de lectura para analizar patrones de texto

def leer_datos(ruta="datos.txt"):                   #Función para leer lod datos desde el archivo
    datos = []                                      #Lista vacia para almacenar los datos
    with open(ruta, "r", encoding="utf-8") as f:    #Abre el archivo en la ruta y lo lee en UTF-8
        for linea in f:                             #Loop para recorrer cada linea del archivo
            linea = linea.strip().rstrip(",")       #Eliminamos los saltos de linea y la coma que seapra a los datos en el texto
            if linea:                               #Condicional para la linea
                m = re.match(r'\{"id":\s*(\d+),\s*"nombre":\s*"([^"]+)",\s*"promedio":\s*([\d.]+)\}', linea)    #Variable que contiene el match según el formato con lo leido en la linea extraida del texto
                if m:                               #Condicional para la variable m si se verifica el match
                    datos.append({                  #Agrega los datos como diccionario a la lista de datos
                        "id": int(m.group(1)),
                        "nombre": m.group(2),
                        "promedio": float(m.group(3))
                    })
    return datos             #entrega la lista o arreglo con los datos
```

```python
import ast   #Lista doblemente ligada con datos aleatorios

class Nodo:                              #Crea la clase nodo
    def __init__(self, datos):           #es el cosntructor d ela clase, inicia la clase nodo con los datos
        self.datos = datos               #El valor del dato, en este caso, el diccionario
        self.anterior = None             #puntero al nodo anteiror
        self.siguiente = None            #puntero al nodo siguiente

class ListaDoblementeLigada:             #Crea la clase para lista doblemente ligada
    def __init__(self):                  #Constructor que incia la lisat vacia
        self.cabeza = None               #Puntero al nodo al principio de la lista
        self.cola = None                 #Puntero al nodo final d ela lista

    def agregar(self, datos):            #Función para agregar datos a la lista
        nuevo = Nodo(datos)              #Crea un nuevo nodo con los datos
        if self.cabeza is None:          #Si no hay nodo al principio crea el nodo cabeza como el nodo nuevo creado anteriormente, el cual tambien es a su vez creado como el nodo cola
            self.cabeza = nuevo
            self.cola = nuevo
        else:                            #Si ya existe un nodo en cabeza....
            nuevo.anterior = self.cola   #Se actualiza el puntero anteiror de la cabeza como el nuevo nodo
            self.cola.siguiente = nuevo  #...se actualiza el puntero al nodo siguiente de la cabeza como el nuevo nodo
            self.cola = nuevo            #...se establece el nuevo nodo como la cola

    def mostrar(self, limite=5):         #Funcion para mostrar la lista
        actual = self.cabeza             #Nombra una variable con el comienzo de la lista
        count = 0                        #Contador para le while
        while actual and count < limite: #While para limitar cuantos datos se van a mostrar segun lo establecido en la variable limite
            print(actual.datos)          #imprime el primer nodo de la lista
            actual = actual.siguiente    #actualiza al siguiente nodo
            count += 1                   #aumenta el contador en 1

lista_a = ListaDoblementeLigada()        #Variable con la lista
for dato in leer_datos():                #Loop para recorrer la lista de datos
    lista_a.agregar(dato)                #agrega el dato a la lista creada

print("lista_a cargada. Primeros 5 nodos:")    #imprime la un mensaje sobre lo que se va a mostrar
lista_a.mostrar()                              #muestra los primeros 5 nodos de la lista
```

```python
import ast     #Arbol ABB datos ALEATORIOS

class NodoArbol:                               #Crea la clase nodo del arbol
    def __init__(self, datos):                 #Constructor de la clase
        self.datos = datos                     #El valor del dato, en este caso, el diccionario
        self.izquierda = None                  #Puntero al nodo del hijo izquierdo del arbol
        self.derecha = None                    #Puntero al nodo del hijo derecho del arbol

class ArbolABB:                                #Crea la clase para el arbol ABB
    def __init__(self):                        #Constructor que incia la lisat vacia
        self.raiz = None                       #Puntero a la raiz del arbol

    def insertar(self, datos):                 #Función para insertar datos en el arbol
        if self.raiz is None:                  #Si no hay nodo en la raiz...
            self.raiz = NodoArbol(datos)       #...EStablece el nodo actual como nodo raiz
            return                             #Devuelve el nodo raiz
        # iterativo en vez de recursivo → sin límite de profundidad
        actual = self.raiz                     #Nombra una variable con el comienzo de la lista
        while True:                            #Loop para verificar id
            if datos["id"] < actual.datos["id"]:      #Si el ID a insertar es menor al ID del nodo actual...
                if actual.izquierda is None:          #...verificamos si el puntero al hijo izquierdo esta vacio...
                    actual.izquierda = NodoArbol(datos)    # Si está vacío, insertamos el nuevo nodo ahí
                    break                             #salimos del loop
                else:                                 #Si no esta vacio...
                    actual = actual.izquierda         #...bajamos a ese hijo y repetimos la comparación
            else:                                     #Si es mayor al id del nodo actual...
                if actual.derecha is None:            #...verificamos si el puntero al hijo derecho esta vacio...
                    actual.derecha = NodoArbol(datos)      # Si está vacío, insertamos el nuevo nodo ahí
                    break                             #salimos del loop
                else:                                 #Si no esta vacio...
                    actual = actual.derecha           #...bajamos a ese hijo y repetimos la comparación

    def en_orden(self, nodo, limite=5, contador=[0]): #Funcion para mostrar el arbol
        if nodo is None or contador[0] >= limite:     #Condicional en base al contador y el limite a ser mostrado
            return                                    #cerramos el condicional
        self.en_orden(nodo.izquierda, limite, contador)  #llamada recursiva de la funcion para el nodo izquierdo
        if contador[0] < limite:                      #Condicional para mostrar
            print(nodo.datos)                         #imprime los datos en el nodo
            contador[0] += 1                          #incrementa el contador
        self.en_orden(nodo.derecha, limite, contador) #llamada recursiva de la funcion para el nodo derecho

arbol_abb_a = ArbolABB()                              #variable con la funcion
for dato in leer_datos():                             #loop para agregar los datos
    arbol_abb_a.insertar(dato)                        #entrada de los datos al arbol

print("arbol_abb_a cargado. Primeros 5 en orden:")    #imprime mensaje de lo que se va a mostrar
arbol_abb_a.en_orden(arbol_abb_a.raiz)                #primero 5 datos en orden
```

```python
import ast      # Librería para interpretar estructuras de datos en texto (no se usa directamente aquí)

ORDEN = 100     # Máximo de claves por nodo, determina cuántos datos o hijos puede tener cada nodo

class NodoBMas:                               # Crea la clase nodo del árbol B+
    def __init__(self, es_hoja=False):        # Constructor: por defecto el nodo es interno, no hoja
        self.es_hoja = es_hoja                # True si el nodo es hoja (guarda datos), False si es interno (guarda referencias)
        self.claves = []                      # Lista de IDs que guía la navegación del árbol
        self.datos = []                       # Lista de diccionarios completos (SOLO se usa en nodos hoja)
        self.hijos = []                       # Lista de punteros a nodos hijos (SOLO se usa en nodos internos)
        self.siguiente = None                 # Puntero al siguiente nodo hoja (SOLO en hojas, para recorrerlas en orden)

class ArbolBMas:                              # Crea la clase del árbol B+
    def __init__(self, orden=ORDEN):          # Constructor: inicializa el árbol con una hoja vacía como raíz
        self.orden = orden                    # Guardamos el orden máximo del árbol
        self.raiz = NodoBMas(es_hoja=True)    # La raíz empieza siendo una hoja vacía

    def insertar(self, dato):                         # Función principal para insertar un dato en el árbol
        id_clave = dato["id"]                         # Extraemos el ID que usaremos como clave de navegación
        resultado = self._insertar_en_nodo(self.raiz, id_clave, dato)  # Intentamos insertar desde la raíz
        if resultado:                                 # Si la raíz se dividió, nos devuelve la clave que sube y el nodo nuevo
            clave_subida, nodo_derecho = resultado    # Desempacamos la clave que sube y el nodo derecho nuevo
            nueva_raiz = NodoBMas(es_hoja=False)      # Creamos una nueva raíz interna
            nueva_raiz.claves = [clave_subida]        # La nueva raíz guarda la clave que subió
            nueva_raiz.hijos = [self.raiz, nodo_derecho]  # Sus dos hijos son la raíz anterior y el nodo nuevo
            self.raiz = nueva_raiz                    # Actualizamos la raíz del árbol

    def _insertar_en_nodo(self, nodo, clave, dato):   # Función recursiva para insertar en el nodo correcto
        if nodo.es_hoja:                              # Si es hoja, aquí es donde se inserta el dato
            i = 0
            while i < len(nodo.claves) and clave > nodo.claves[i]:  # Buscamos la posición correcta para mantener el orden
                i += 1
            nodo.claves.insert(i, clave)              # Insertamos la clave en la posición encontrada
            nodo.datos.insert(i, dato)                # Insertamos el diccionario en la misma posición
            if len(nodo.claves) >= self.orden:        # Si la hoja se llenó (alcanzó el orden máximo)...
                return self._dividir_hoja(nodo)       # ...la dividimos y retornamos la clave que sube
            return None                               # Si no se llenó, no hay nada que retornar
        else:                                         # Si es nodo interno, tenemos que bajar al hijo correcto
            i = 0
            while i < len(nodo.claves) and clave >= nodo.claves[i]:  # Buscamos por cuál hijo bajar
                i += 1
            resultado = self._insertar_en_nodo(nodo.hijos[i], clave, dato)  # Bajamos recursivamente
            if resultado:                             # Si el hijo se dividió, nos devuelve la clave que sube
                clave_subida, nodo_derecho = resultado        # Desempacamos la clave y el nodo nuevo
                nodo.claves.insert(i, clave_subida)           # Incorporamos la clave que subió
                nodo.hijos.insert(i + 1, nodo_derecho)        # Agregamos el nuevo nodo como hijo
                if len(nodo.claves) >= self.orden:            # Si este nodo interno también se llenó...
                    return self._dividir_interno(nodo)        # ...lo dividimos también
            return None                               # Si no hubo división, no retornamos nada

    def _dividir_hoja(self, hoja):                    # Función para dividir una hoja llena en dos
        mitad = len(hoja.claves) // 2                 # Calculamos el punto de división
        nueva_hoja = NodoBMas(es_hoja=True)           # Creamos la nueva hoja que recibirá la mitad derecha
        nueva_hoja.claves = hoja.claves[mitad:]       # La nueva hoja se queda con las claves de la mitad derecha
        nueva_hoja.datos = hoja.datos[mitad:]         # La nueva hoja se queda con los datos de la mitad derecha
        nueva_hoja.siguiente = hoja.siguiente         # La nueva hoja hereda el puntero siguiente de la hoja original
        hoja.claves = hoja.claves[:mitad]             # La hoja original se queda solo con la mitad izquierda
        hoja.datos = hoja.datos[:mitad]               # Lo mismo con sus datos
        hoja.siguiente = nueva_hoja                   # La hoja original ahora apunta a la nueva hoja
        return (nueva_hoja.claves[0], nueva_hoja)     # Retornamos la primera clave de la nueva hoja para que suba al padre

    def _dividir_interno(self, nodo):                 # Función para dividir un nodo interno lleno en dos
        mitad = len(nodo.claves) // 2                 # Calculamos el punto de división
        clave_subida = nodo.claves[mitad]             # La clave del medio sube al padre (no queda en ningún hijo)
        nuevo_nodo = NodoBMas(es_hoja=False)          # Creamos el nuevo nodo interno que recibirá la mitad derecha
        nuevo_nodo.claves = nodo.claves[mitad + 1:]   # El nuevo nodo se queda con las claves de la derecha
        nuevo_nodo.hijos = nodo.hijos[mitad + 1:]     # El nuevo nodo se queda con los hijos de la derecha
        nodo.claves = nodo.claves[:mitad]             # El nodo original se queda con las claves de la izquierda
        nodo.hijos = nodo.hijos[:mitad + 1]           # El nodo original se queda con los hijos de la izquierda
        return (clave_subida, nuevo_nodo)             # Retornamos la clave que sube y el nuevo nodo

    def recorrer_hojas(self, limite=5):               # Función para mostrar los primeros n datos del árbol
        nodo = self.raiz
        while not nodo.es_hoja:                       # Bajamos hasta la primera hoja (la de menor ID)
            nodo = nodo.hijos[0]
        count = 0
        while nodo and count < limite:                # Recorremos las hojas enlazadas hasta alcanzar el límite
            for dato in nodo.datos:                   # Iteramos sobre los datos de cada hoja
                if count >= limite:                   # Si ya alcanzamos el límite, paramos
                    break
                print(dato)                           # Imprimimos el diccionario del estudiante
                count += 1                            # Contamos el dato impreso
            nodo = nodo.siguiente                     # Avanzamos a la siguiente hoja enlazada

    def info_arbol(self):                             # Función para mostrar información estructural del árbol
        nivel = 0
        nodo = self.raiz
        while not nodo.es_hoja:                       # Contamos niveles bajando siempre por el hijo más a la izquierda
            nivel += 1
            nodo = nodo.hijos[0]
        print(f"Niveles del árbol: {nivel + 1} (incluyendo hojas)")   # Mostramos el total de niveles
        print(f"Nodos hijos de la raíz: {len(self.raiz.hijos)}")      # Mostramos cuántos hijos tiene la raíz

# --- Cargar datos e insertarlos en el árbol B+ desordenado ---
arbol_bmas_a = ArbolBMas()                # Creamos el árbol vacío
for dato in leer_datos():                 # Leemos cada diccionario del archivo tal como viene
    arbol_bmas_a.insertar(dato)           # Lo insertamos en el árbol (se balancea automáticamente)

print("=== arbol_bmas_a (desordenado) ===")
arbol_bmas_a.info_arbol()                 # Mostramos la estructura del árbol
arbol_bmas_a.recorrer_hojas()             # Mostramos los primeros 5 datos
```

```python
import time
import random
from faker import Faker
import pandas as pd

fake = Faker('es_ES')

# ----------------------------------------------------------------
# BÚSQUEDA INDIVIDUAL
# ----------------------------------------------------------------
def buscar_lista(lista, id_buscar):         # Función para buscar un ID en la lista doble
    actual = lista.cabeza                   # Empezamos desde el primer nodo
    while actual:                           # Mientras haya nodos por recorrer
        if actual.datos["id"] == id_buscar: # Si el ID del nodo actual coincide con el buscado
            return actual.datos             # Retornamos el diccionario encontrado
        actual = actual.siguiente           # Si no coincide, avanzamos al siguiente nodo
    return None                             # Si no se encontró, retornamos None

def buscar_abb(arbol, id_buscar):           # Función para buscar un ID en el árbol ABB
    nodo = arbol.raiz                       # Empezamos desde la raíz
    while nodo is not None:                 # Mientras haya nodos por recorrer
        if id_buscar == nodo.datos["id"]:   # Si el ID coincide, lo encontramos
            return nodo.datos
        elif id_buscar < nodo.datos["id"]:  # Si el buscado es menor, bajamos a la izquierda
            nodo = nodo.izquierda
        else:                               # Si es mayor, bajamos a la derecha
            nodo = nodo.derecha
    return None                             # Si salimos del while, el ID no existe

def buscar_bmas(arbol, id_buscar):          # Función para buscar un ID en el árbol B+
    nodo = arbol.raiz                       # Empezamos desde la raíz
    while not nodo.es_hoja:                 # Bajamos por nodos internos hasta llegar a una hoja
        i = 0
        while i < len(nodo.claves) and id_buscar >= nodo.claves[i]:  # Buscamos el hijo correcto
            i += 1
        nodo = nodo.hijos[i]                # Bajamos al hijo correspondiente
    for dato in nodo.datos:                 # Ya en la hoja, buscamos el ID entre los datos
        if dato["id"] == id_buscar:
            return dato                     # Retornamos el diccionario si lo encontramos
    return None                             # Si no está en la hoja, no existe

# ----------------------------------------------------------------
# LISTADO ORDENADO POR ID
# ----------------------------------------------------------------
def listar_lista(lista):                    # Función para listar la lista doble ordenada por ID
    resultado = []                          # Lista auxiliar para almacenar los datos
    actual = lista.cabeza                   # Empezamos desde el primer nodo
    while actual:                           # Recorremos todos los nodos
        resultado.append(actual.datos)      # Agregamos el diccionario de cada nodo
        actual = actual.siguiente           # Avanzamos al siguiente nodo
    return sorted(resultado, key=lambda x: x["id"])  # Ordenamos por ID antes de retornar

def listar_abb(arbol):                      # Función para listar el ABB ordenado por ID
    resultado = []                          # Lista auxiliar para almacenar los datos
    def _en_orden(nodo):                    # Recorrido en orden: izquierda → actual → derecha
        if nodo is None:                    # Caso base: nodo vacío, no hacemos nada
            return
        _en_orden(nodo.izquierda)           # Primero recorremos todos los menores
        resultado.append(nodo.datos)        # Luego agregamos el nodo actual
        _en_orden(nodo.derecha)             # Finalmente recorremos todos los mayores
    _en_orden(arbol.raiz)                   # Iniciamos desde la raíz
    return resultado                        # El ABB en orden ya devuelve ordenado por ID

def listar_bmas(arbol):                     # Función para listar el B+ ordenado por ID
    resultado = []                          # Lista auxiliar para almacenar los datos
    nodo = arbol.raiz
    while not nodo.es_hoja:                 # Bajamos hasta la primera hoja (la de menor ID)
        nodo = nodo.hijos[0]
    while nodo:                             # Recorremos las hojas enlazadas de izquierda a derecha
        resultado.extend(nodo.datos)        # Agregamos todos los datos de la hoja actual
        nodo = nodo.siguiente               # Avanzamos a la siguiente hoja
    return resultado                        # Las hojas ya están ordenadas por ID

# ----------------------------------------------------------------
# INSERCIÓN
# ----------------------------------------------------------------
def generar_nuevo_dato():                           # Genera un diccionario con datos aleatorios
    return {
        "id": random.randint(100000, 999999),       # ID fuera del rango original (0-99999)
        "nombre": fake.first_name(),                # Nombre aleatorio en español
        "promedio": round(random.uniform(0.0, 10.0), 1)  # Promedio aleatorio con 1 decimal
    }

# ----------------------------------------------------------------
# UTILIDADES
# ----------------------------------------------------------------
def medir(fn, *args):                       # Mide el tiempo de ejecución de una función
    inicio = time.perf_counter()            # Registramos el tiempo antes de ejecutar
    resultado = fn(*args)                   # Ejecutamos la función con sus argumentos
    fin = time.perf_counter()              # Registramos el tiempo después de ejecutar
    return resultado, round((fin - inicio) * 1000, 4)  # Retornamos resultado y tiempo en ms

def mostrar_tabla(operacion, tiempos):      # Muestra una tabla comparativa de tiempos
    df = pd.DataFrame({
        "Estructura": [                     # Nombres de cada estructura
            "Lista doble  — desordenada",
            "Lista doble  — ordenada",
            "ABB          — desordenado",
            "ABB          — ordenado (degenerado)",
            "B+           — desordenado",
            "B+           — ordenado",
        ],
        f"Tiempo {operacion} (ms)": tiempos # Columna con los tiempos medidos
    })
    print(df.to_string(index=False))        # Imprimimos la tabla sin índice de filas

# ----------------------------------------------------------------
# BÚSQUEDA MASIVA (nueva función)
# ----------------------------------------------------------------
def busqueda_masiva(n_busquedas):
    # Generamos todos los IDs a buscar de una vez (aleatorios entre 0 y 99999)
    ids_a_buscar = [random.randint(0, 99999) for _ in range(n_busquedas)]

    # Lista con las 6 estructuras y sus funciones de búsqueda
    estructuras = [
        ("Lista doble  — desordenada",       buscar_lista, lista_a),
        ("Lista doble  — ordenada",          buscar_lista, lista_b),
        ("ABB          — desordenado",       buscar_abb,   arbol_abb_a),
        ("ABB          — ordenado (degen.)", buscar_abb,   arbol_abb_b),
        ("B+           — desordenado",       buscar_bmas,  arbol_bmas_a),
        ("B+           — ordenado",          buscar_bmas,  arbol_bmas_b),
    ]

    tiempos_promedio = []                   # Lista para guardar el tiempo promedio de cada estructura

    for nombre, fn, est in estructuras:     # Iteramos sobre cada estructura
        tiempos = []                        # Lista para guardar los tiempos de cada búsqueda

        for id_buscar in ids_a_buscar:      # Realizamos cada una de las n búsquedas
            _, t = medir(fn, est, id_buscar)# Medimos el tiempo de esa búsqueda individual
            tiempos.append(t)               # Guardamos el tiempo

        promedio = round(sum(tiempos) / len(tiempos), 4)  # Calculamos el tiempo promedio
        tiempos_promedio.append(promedio)   # Guardamos el promedio de esta estructura

    # Mostramos la tabla con los tiempos promedio
    df = pd.DataFrame({
        "Estructura": [e[0] for e in estructuras],              # Nombre de cada estructura
        f"Promedio por búsqueda ({n_busquedas} búsquedas) (ms)": tiempos_promedio  # Tiempos promedio
    })
    print(f"\n=== RESULTADOS: {n_busquedas} búsquedas aleatorias ===")
    print(df.to_string(index=False))        # Imprimimos la tabla sin índice

# ----------------------------------------------------------------
# MENÚ
# ----------------------------------------------------------------
def menu():
    while True:
        print("\n" + "=" * 45)
        print("         MENÚ DE OPERACIONES")
        print("=" * 45)
        print("  1. Buscar estudiante por ID")
        print("  2. Insertar nuevo estudiante (random)")
        print("  3. Listar todos ordenados por ID")
        print("  4. Prueba de rendimiento por búsquedas")  # Nueva opción
        print("  0. Salir")
        print("=" * 45)
        opcion = input("Elige una opción: ").strip()

        # ── BUSCAR ──────────────────────────────────────
        if opcion == "1":
            try:
                id_buscar = int(input("ID a buscar (0-99999): ").strip())  # Pedimos el ID
            except ValueError:
                print("⚠ ID inválido, ingresa un número.")
                continue

            tiempos = []
            resultados = []
            for fn, est in [                # Ejecutamos la búsqueda en las 6 estructuras
                (buscar_lista, lista_a),
                (buscar_lista, lista_b),
                (buscar_abb,   arbol_abb_a),
                (buscar_abb,   arbol_abb_b),
                (buscar_bmas,  arbol_bmas_a),
                (buscar_bmas,  arbol_bmas_b),
            ]:
                res, t = medir(fn, est, id_buscar)  # Medimos el tiempo de cada búsqueda
                resultados.append(res)
                tiempos.append(t)

            encontrado = next((r for r in resultados if r), None)  # Tomamos el primer resultado encontrado
            if encontrado:
                print(f"\nEstudiante encontrado: {encontrado}")
            else:
                print(f"\nID {id_buscar} no encontrado.")

            mostrar_tabla("búsqueda", tiempos)  # Mostramos tabla comparativa

        # ── INSERTAR ────────────────────────────────────
        elif opcion == "2":
            nuevo_dato = generar_nuevo_dato()           # Generamos un dato aleatorio
            print(f"\nInsertando: {nuevo_dato}")

            tiempos = []
            for est in [lista_a, lista_b, arbol_abb_a, arbol_abb_b, arbol_bmas_a, arbol_bmas_b]:
                _, t = medir(est.insertar, nuevo_dato.copy())  # Insertamos en cada estructura
                tiempos.append(t)

            mostrar_tabla("inserción", tiempos)         # Mostramos tabla comparativa

        # ── LISTAR ──────────────────────────────────────
        elif opcion == "3":
            tiempos = []
            for fn, est in [                # Ejecutamos el listado en las 6 estructuras
                (listar_lista, lista_a),
                (listar_lista, lista_b),
                (listar_abb,   arbol_abb_a),
                (listar_abb,   arbol_abb_b),
                (listar_bmas,  arbol_bmas_a),
                (listar_bmas,  arbol_bmas_b),
            ]:
                _, t = medir(fn, est)       # Medimos el tiempo de cada listado
                tiempos.append(t)

            mostrar_tabla("listado", tiempos)           # Mostramos tabla comparativa
            print("(Listado completo omitido para no saturar la salida)")

        # ── PRUEBA DE RENDIMIENTO ────────────────────────
        elif opcion == "4":
            print("\n¿Cuántas búsquedas aleatorias quieres realizar?")
            print("  1. 100 búsquedas")
            print("  2. 1,000 búsquedas")
            print("  3. 2,000 búsquedas")
            print("  4. 4,000 búsquedas")
            sub_opcion = input("Elige una opción: ").strip()

            # Mapeamos la opción al número de búsquedas correspondiente
            opciones_busquedas = {"1": 100, "2": 1000, "3": 2000, "4": 4000}

            if sub_opcion in opciones_busquedas:                    # Si la opción es válida
                n = opciones_busquedas[sub_opcion]                  # Obtenemos el número de búsquedas
                print(f"\nEjecutando {n} búsquedas aleatorias...")
                busqueda_masiva(n)                                  # Ejecutamos la prueba
            else:
                print("⚠ Opción no válida.")                        # Si no es válida, avisamos

        # ── SALIR ───────────────────────────────────────
        elif opcion == "0":
            print("Saliendo...")
            break
        else:
            print("⚠ Opción no válida.")        # Si la opción no existe, avisamos

menu()                                          # Ejecutamos el menú
```
