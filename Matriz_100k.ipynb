```python
import os

filename = "matriz_100k_byte.bin"

N = 100_000
row_size = N                      # 1 byte por celda
block_rows = 100                  # escribir 100 filas por bloque

with open(filename, "wb") as f:
    for i in range(0, N, block_rows):
        rows_to_write = min(block_rows, N - i)

        # ejemplo: fila alternando 0 y 1
        row = bytes([ (j % 2) for j in range(N) ])

        block = row * rows_to_write
        f.write(block)

print("Archivo creado.")
print("Tamaño:", os.path.getsize(filename))
```

```python
filename = "matriz_100k_byte.bin"
num_bytes_to_read = 100

try:
    with open(filename, "rb") as f:
        first_bytes = f.read(num_bytes_to_read)

    print(f"First {len(first_bytes)} bytes of '{filename}':")
    print(first_bytes.hex())

except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```
