```python
import os                                              # Librería para interactuar con el sistema de archivos

filename = "matriz_100k_byte.bin"                      # Nombre del archivo binario que se va a crear

N = 100_000                                            # Tamaño de la matriz: 100,000 filas y 100,000 columnas
row_size = N                                           # Tamaño de cada fila en bytes (1 byte por celda)
block_rows = 100                                       # Cantidad de filas que se escriben por bloque para no cargar todo en memoria

with open(filename, "wb") as f:                        # Abrimos el archivo en modo escritura binaria (wb = write binary)
    for i in range(0, N, block_rows):                  # Iteramos de 0 a N avanzando de block_rows en block_rows
        rows_to_write = min(block_rows, N - i)         # Calculamos cuántas filas quedan por escribir (evita pasarse del límite en la última vuelta)

        row = bytes([ (j % 2) for j in range(N) ])     # Creamos una fila de N bytes alternando 0 y 1 (0,1,0,1,0,1...)
        block = row * rows_to_write                    # Repetimos la fila tantas veces como filas hay en el bloque actual
        f.write(block)                                 # Escribimos el bloque completo en el archivo binario

print("Archivo creado.")
print("Tamaño:", os.path.getsize(filename))            # Imprimimos el tamaño final del archivo en bytes
```

```python
filename = "matriz_100k_byte.bin"                              # Nombre del archivo binario que vamos a leer
num_bytes_to_read = 100                                        # Cantidad de bytes que queremos leer del archivo

try:                                                           # Intentamos ejecutar el bloque, capturando posibles errores
    with open(filename, "rb") as f:                            # Abrimos el archivo en modo lectura binaria (rb = read binary)
        first_bytes = f.read(num_bytes_to_read)                # Leemos solo los primeros 100 bytes del archivo

    print(f"First {len(first_bytes)} bytes of '{filename}':")  # Imprimimos cuántos bytes leímos y de qué archivo
    print(first_bytes.hex())                                   # Convertimos los bytes a hexadecimal y los imprimimos (ej: 0101010101...)

except FileNotFoundError:                                      # Si el archivo no existe, capturamos ese error específico
    print(f"Error: File '{filename}' not found.")              # Informamos que no se encontró el archivo

except Exception as e:                                         # Capturamos cualquier otro error inesperado
    print(f"An error occurred: {e}")                           # Imprimimos el mensaje del error para saber qué salió mal
```
